```python
#Feature Engineering 
#1- Character length in questions 
def length_func(row) :
    x= len(row)
        
    return x 

new_df['q1_len'] = new_df['question1'].apply(length_func)

new_df['q2_len'] = new_df['question2'].apply(lambda x : len(x))
```

```python
new_df 
```

```python
#2- Word Count or no of words in Question 1 and Question 2

    
new_df['q1_num_words'] = new_df['question1'].apply(lambda x : len(str(x).split(" ")))

new_df['q2_num_words'] = new_df['question2'].apply(lambda x : len(str(x).split(" ")))
```

```python
#We can also find it using this way 
def word_func(row) :
    w1 = set(map(lambda x : x.lower().strip() , row['question1'].split(" ")))
    w2 = set(map(lambda x : x.lower().strip() , row['question2'].split(" ")))
    return len(w1& w2 )

new_df['word_common'] = new_df.apply(word_func , axis =1 )
```

```python
#4- Total word count 

def total_words(row) :
    w1 = set(map(lambda x : x.lower().strip() , row['question1'].split(" ")))
    w2 = set(map(lambda x : x.lower().strip() , row['question2'].split(" ")))
    return (len(w1) + len(w2))

new_df['word_total'] = new_df.apply(total_words , axis = 1 )
```

```python
new_df 
```

```python
new_df['words_share'] = round(new_df['word_common'] / new_df['word_total'] , 2)
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
cv = CountVectorizer(max_features = 1500)
question_list = new_df['question1'].to_list() + new_df['question2'].to_list()
q1_arr , q2_arr = np.vsplit(cv.fit_transform(question_list).toarray() , 2)
```

```python
ques_df = new_df[['question1','question2']]
ques_df.head()
```

```python
temp1_df = pd.DataFrame(q1_arr , index =ques_df.index )
temp2_df = pd.DataFrame(q2_arr , index = ques_df.index)
temp_df = pd.concat([temp1_df , temp2_df] , axis = 1)
```

```python
temp_df.shape
temp_df.head()
```

```python
final_df = new_df.drop(['id' ,'qid1' ,'qid2' , 'question1' ,'question2'] , axis = 1 )
```

```python
final_df 
```

```python
final_df = pd.concat([final_df , temp_df] , axis =1 )
```

```python
final_df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(final_df.iloc[:,1:].values,final_df.iloc[:,0].values,test_size=0.2,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier()
rf.fit(X_train , y_train)
y_predict = rf.predict(X_test)
accuracy_score(y_test , y_predict)
```
